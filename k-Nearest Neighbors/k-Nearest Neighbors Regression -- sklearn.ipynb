# Nearest Neighbors Regression -- sklearn

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import neighbors

rng = np.random.default_rng(10)

X_train = np.sort(5 * rng.random((50, 1)), axis=0)
X_test = np.linspace(0, 5, 500)[:, np.newaxis]
y = np.sin(X_train).ravel()

y[::5] += 1 * (0.5 - rng.random(10))
```

```python
n_neighbors = 5 

for i, weights in enumerate(["uniform", "distance"]):
    knn = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
    y_ = knn.fit(X_train, y).predict(X_test) 

    plt.subplot(2, 1, i + 1)
    plt.scatter(X_train, y, color="darkorange", label="data")
    plt.plot(X_test, y_, color="navy", label="prediction")
    plt.axis("tight")
    plt.legend()
    plt.title("KNeighborsRegressor (k = %i, weights = '%s')" % (n_neighbors, weights))

plt.tight_layout() 
plt.show()
```
